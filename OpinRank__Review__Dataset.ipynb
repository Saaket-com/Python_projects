```python
import numpy as np

print("### Simulating OpinRank Review Dataset with NumPy ###")

# Set random seed for reproducibility
np.random.seed(42)

num_reviews = 1000

# Simulate data for key features
# product_id: 100 unique products
product_ids = np.random.randint(1, 101, num_reviews)
# user_id: 500 unique users
user_ids = np.random.randint(1, 501, num_reviews)
# rating: 1.0 to 5.0 (rounded to 1 decimal place)
ratings = np.round(np.random.uniform(1.0, 5.0, num_reviews), 1)
# review_length: 20 to 500 words
review_lengths = np.random.randint(20, 501, num_reviews)
# helpfulness_votes: 0 to 200
helpfulness_votes = np.random.randint(0, 201, num_reviews)
# sentiment_score: -1.0 to 1.0 (rounded to 2 decimal places)
sentiment_scores = np.round(np.random.uniform(-1.0, 1.0, num_reviews), 2)
# day_of_week: 0=Monday to 6=Sunday
day_of_week = np.random.randint(0, 7, num_reviews)

# Combine into a dictionary for easier access
data = {
    'product_id': product_ids,
    'user_id': user_ids,
    'rating': ratings,
    'review_length': review_lengths,
    'helpfulness_votes': helpfulness_votes,
    'sentiment_score': sentiment_scores,
    'day_of_week': day_of_week
}

print(f"Simulated {num_reviews} reviews with the following features:\n")
for key, value in data.items():
    print(f"  {key}: first 5 values {value[:5]}, dtype {value.dtype}")

print("\n---\n")
print("### 20 Problem Statements and NumPy Solutions ###\n")

# Problem 1
print("1. Problem Statement: Calculate the overall average rating of all reviews.")
solution_1 = np.mean(data['rating'])
print(f"   Solution: {solution_1:.2f}\n")

# Problem 2
print("2. Problem Statement: Find the minimum and maximum sentiment scores observed.")
solution_2_min = np.min(data['sentiment_score'])
solution_2_max = np.max(data['sentiment_score'])
print(f"   Solution: Min Sentiment Score = {solution_2_min:.2f}, Max Sentiment Score = {solution_2_max:.2f}\n")

# Problem 3
print("3. Problem Statement: Determine the total number of reviews.")
solution_3 = len(data['rating'])
print(f"   Solution: {solution_3} reviews\n")

# Problem 4
print("4. Problem Statement: Count how many reviews received a perfect 5.0 rating.")
solution_4 = np.sum(data['rating'] == 5.0)
print(f"   Solution: {solution_4} reviews\n")

# Problem 5
print("5. Problem Statement: Calculate the standard deviation of review lengths.")
solution_5 = np.std(data['review_length'])
print(f"   Solution: {solution_5:.2f}\n")

# Problem 6
print("6. Problem Statement: Identify the product IDs that have at least one 1.0-star rating.")
solution_6 = np.unique(data['product_id'][data['rating'] == 1.0])
print(f"   Solution: {solution_6}\n")

# Problem 7
print("7. Problem Statement: Find the average helpfulness votes for reviews with a positive sentiment score (sentiment_score > 0).")
positive_sentiment_helpfulness = data['helpfulness_votes'][data['sentiment_score'] > 0]
solution_7 = np.mean(positive_sentiment_helpfulness) if positive_sentiment_helpfulness.size > 0 else 0
print(f"   Solution: {solution_7:.2f}\n")

# Problem 8
print("8. Problem Statement: Determine the top 5 most frequently occurring review lengths.")
unique_lengths, counts = np.unique(data['review_length'], return_counts=True)
top_5_indices = np.argsort(counts)[-5:][::-1]
solution_8_lengths = unique_lengths[top_5_indices]
solution_8_counts = counts[top_5_indices]
print(f"   Solution: Top 5 lengths and their counts: {list(zip(solution_8_lengths, solution_8_counts))}\n")

# Problem 9
print("9. Problem Statement: Calculate the percentage of reviews that are considered \"short\" (less than 50 words).")
solution_9 = np.sum(data['review_length'] < 50) / len(data['review_length']) * 100
print(f"   Solution: {solution_9:.2f}%\n")

# Problem 10
print("10. Problem Statement: Find the average rating for each day of the week (0=Mon, 6=Sun).")
solution_10 = [np.mean(data['rating'][data['day_of_week'] == day]) for day in range(7)]
print(f"   Solution: {solution_10}\n")

# Problem 11
print("11. Problem Statement: Identify all unique user IDs present in the dataset.")
solution_11 = np.unique(data['user_id'])
print(f"   Solution: First 10 unique user IDs: {solution_11[:10]}\n")

# Problem 12
print("12. Problem Statement: Get all review lengths for reviews that have 0 helpfulness votes.")
solution_12 = data['review_length'][data['helpfulness_votes'] == 0]
print(f"   Solution: First 10 review lengths with 0 helpfulness votes: {solution_12[:10]}\n")

# Problem 13
print("13. Problem Statement: Calculate the median rating.")
solution_13 = np.median(data['rating'])
print(f"   Solution: {solution_13:.2f}\n")

# Problem 14
print("14. Problem Statement: Find the product ID that has the highest average review length.")
unique_products = np.unique(data['product_id'])
avg_len_per_product = [np.mean(data['review_length'][data['product_id'] == pid]) for pid in unique_products]
solution_14 = unique_products[np.argmax(avg_len_per_product)]
print(f"   Solution: Product ID {solution_14}\n")

# Problem 15
print("15. Problem Statement: Count the number of reviews where the sentiment score is negative (sentiment_score < 0).")
solution_15 = np.sum(data['sentiment_score'] < 0)
print(f"   Solution: {solution_15} reviews\n")

# Problem 16
print("16. Problem Statement: Extract all ratings for reviews that were written on a Sunday (day_of_week = 6).")
solution_16 = data['rating'][data['day_of_week'] == 6]
print(f"   Solution: First 10 ratings from Sunday reviews: {solution_16[:10]}\n")

# Problem 17
print("17. Problem Statement: Determine the product ID with the most reviews.")
unique_products, counts = np.unique(data['product_id'], return_counts=True)
solution_17 = unique_products[np.argmax(counts)]
print(f"   Solution: Product ID {solution_17}\n")

# Problem 18
print("18. Problem Statement: Calculate the sum of helpfulness votes for all reviews that have a rating of 4.0 or higher.")
solution_18 = np.sum(data['helpfulness_votes'][data['rating'] >= 4.0])
print(f"   Solution: {solution_18}\n")

# Problem 19
print("19. Problem Statement: Find the number of reviews that have both a 5.0-star rating AND a positive sentiment score.")
solution_19 = np.sum((data['rating'] == 5.0) & (data['sentiment_score'] > 0))
print(f"   Solution: {solution_19} reviews\n")

# Problem 20
print("20. Problem Statement: Sort the (conceptual) dataset by helpfulness_votes in descending order and retrieve the top 3 review lengths from this sorted order.")
sorted_indices = np.argsort(data['helpfulness_votes'])[::-1]
solution_20 = data['review_length'][sorted_indices][:3]
print(f"   Solution: Top 3 review lengths by helpfulness votes: {solution_20}\n")
```
